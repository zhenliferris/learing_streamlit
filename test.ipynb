```python
import matplotlib.pyplot as plt


x = list(range(1, 51))  # Updated to generate a list of integers
y = [i**2 for i in x]  # Corrected list comprehension
z = [i**2.2 for i in x]  # Added a list comprehension for z

plt.style.use('seaborn-v0_8-dark')
fig, ax = plt.subplots()  # Fixed the variable name to 'fig' instead of 'fit'
ax.scatter(x, y, c='blue', s=10)
ax.scatter(x, z, c='red', s=10)
ax.fill_between(x, y, z, color='yellow', alpha=0.3)
ax.set_title('Scatter plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```
